## Estructuras de Control (Condicionales y Ciclos)
Haz un programa que ayude a automatizar procesos repetitivos y aplicar reglas de negocio (lógica condicional).

1. Crea un ciclo for que recorra tu lista de registro_ventas (puedes agregar un par de ventas manuales extra al código para hacer la prueba).
2. Dentro del ciclo, aplica un condicional if: Si el total de la venta es mayor a $100000, aplica un descuento del 10%.
3. Imprime el reporte final mostrando el producto, el total original y el total con descuento (si aplica).

```python
##Agregar venta2
registro_ventas.append({"venta2":{
        "producto": "Chaqueta",
        "categoría": "ropa",
        "precio": 150000,
        "cantidad": 3,
        "total": 450000
    }})
print (registro_ventas)
```

```python
##Agregar venta3
registro_ventas.append({"venta3":{
        "producto": "aretes",
        "categoría": "bisutería",
        "precio": 9800,
        "cantidad": 1,
        "total": 9800
    }})
```

```python
print(registro_ventas)
```

```python
##Ciclo con condicional (utilizo el x.values para extraer los totales en el diccionario) (unicamente para saber si aplica descuento o no)
for x in registro_ventas:
    for datos_venta in x.values():
        if datos_venta["total"] > 100000:
            print ("Aplica un descuento del 10%.")
        else:
            print ("No aplica ningún descuento en la compra.")

```

```python
##Reporte final con producto, total base y total con descuento.
for x in registro_ventas:
    for datos_venta in x.values():
        productos = datos_venta ["producto"]
        total_original = datos_venta ["total"]
        if total_original > 100000:
            total_descuento = total_original - (total_original*0.10)
        else:
            total_descuento = total_original
        print (f"{productos} con un valor total de {total_original:.0f} y por total mayor a 100.000, descuento del 10%: {total_descuento:.0f} (valor a pagar)")

```

```python
##Primer error- retormar 0
def calcular_ingreso (precio, cantidad):
    try:
        resultado = precio * cantidad
        return resultado
    except TypeError:
        print ("ERROR: Los valores deben ser numéricos.")
        return 0

```
